```python
# traverse clean_midi
# function to load midi format in np array
maxgroupsz = 0
maxpad = 32112 
allMIDI, labelMIDI = [], []
def processMIDI(src, name):
    try:
        fp = MidiFile(src)
        data = [ [0]*maxpad for i in range(20) ] 
        appendat = [0 for i in range(20)]
        current_group = 0

        for i, track in enumerate(fp.tracks):
            for msg in track:
                if msg.type=='program_change':
                    current_group = instruments[msg.program]
                if msg.type=='note_on':
                    data[current_group][ appendat[current_group] ]=msg.note
                    appendat[current_group] += 1
                    

        for y in range(20):
            global maxgroupsz
            maxgroupsz = max( maxgroupsz, len(data[y]))
        allMIDI.append( np.reshape(np.asarray(data), 20*maxpad) )
        labelMIDI.append(songNames[name])
    except:
        print(src)
    
    #print(np.asarray(data))
pth1 = 'D:\\study\\major project\\scl\\clean_midi'
cnt =0
for folder in os.listdir(pth1):
    for file in os.listdir(os.path.join(pth1, folder)):
        name = file[:-4].lower()
        if name in songNames:    
            processMIDI(pth1+"\\"+folder+"\\"+file, name)
            print(cnt, name)
            cnt+=1

print(maxgroupsz)
```

```python
#total 3922 70%:30% 
X_test=[]
X_train=[]
Y_train=[]
Y_test=[]
i=0
for label in Y:
    if i<2750:
        Y_train.append(label)
    else:
        Y_test.append(label)
    i+=1
i=0
for row in allMIDI:
    if i<2750:
        X_train.append(row)
    else:
        X_test.append(row)
    i+=1   
print(len(X_train))
print(len(X_test))
print(len(Y_train))
print(len(Y_test))
print(Y_train[0])
print(Y_test[0])
allMIDI=None
labelMIDI=None
```

```python
batch_size = 128
num_classes = 2
epochs = 5

# input image dimensions

img_x, img_y = 20,maxpad
```

```python
#list to numpy array conversion
x_train=np.asarray(X_train)
x_test=np.asarray(X_test)
y_train=np.asarray(Y_train)
y_test=np.asarray(Y_test)
x_train = x_train.reshape(x_train.shape[0], img_y, img_x,1)
x_test = x_test.reshape(x_test.shape[0], img_y, img_x,1)
input_shape = (img_y, img_x,1)
```

```python
print(x_train.shape)
print(y_train)
print('done')
```

```python
model = Sequential()
model.add(Conv2D(32, kernel_size=(25, 5), strides=(3,3),
                 activation='relu',
                 padding='same',
                 input_shape=input_shape,
                data_format='channels_last'))

print(model.output_shape)
model.add(MaxPooling2D(pool_size=(2,2),strides=(2,2)))
print(model.output_shape)
model.add(Conv2D(64, kernel_size=(10, 3), strides=(1, 3),

                  activation='relu',
                  padding='same',
                  input_shape=input_shape,
                 data_format='channels_last'))

print(model.output_shape)
model.add(MaxPooling2D(pool_size=(2,1),strides=(2,1)))
print(model.output_shape)
model.add(Flatten())
print(model.output_shape)
#model.add(Dropout(0.2))
#model.add(Dense(1024, actxivation='relu'))
model.add(Dense(1024, activation='relu'))
print(model.output_shape)
model.add(Dense(num_classes, activation='softmax'))
print(model.output_shape)
model.compile(loss=keras.losses.categorical_crossentropy,

              optimizer=keras.optimizers.Adam(),

              metrics=['accuracy'])





class AccuracyHistory(keras.callbacks.Callback):

    def on_train_begin(self, logs={}):

        self.acc = []



    def on_epoch_end(self, batch, logs={}):

        self.acc.append(logs.get('acc'))



history = AccuracyHistory()



model.fit(x_train, y_train,

          batch_size=batch_size,

          epochs=epochs,

          verbose=1,

          validation_data=(x_test, y_test),

          callbacks=[history])

score = model.evaluate(x_test, y_test, verbose=0)

print('Test loss:', score[0])

print('Test accuracy:', score[1])

'''plt.plot(range(1, 11), history.acc)

plt.xlabel('Epochs')

plt.ylabel('Accuracy')

plt.show()    
'''

```
